# Matplotlib

```python
# Courbe simple 
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker="o", color="black", label="y=2x")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Histogramme simple 
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=20, color="skyblue", edgecolor="black")
plt.title("Histogramme - distribution des données")
plt.xlabel("Valeurs")
plt.ylabel("Fréquence")
plt.show()
```

```python
# Ici on a un histogramme avec plusieurs distribution 

data1 = np.random.normal(0, 1, 1000)   # moyenne 0, écart-type 1
data2 = np.random.normal(3, 1.5, 1000) # moyenne 3, écart-type 1.5
data3 = np.random.normal(6, 2, 1000)   # moyenne 6, écart-type 2
data4 = np.random.normal(9, 2.5, 1000) # moyenne 9, écart-type 2.5

plt.hist(data1, bins=30, alpha=1, label="Dist 1")
plt.hist(data2, bins=30, alpha=1, label="Dist 2")
plt.hist(data3, bins=30, alpha=1, label="Dist 3")
plt.hist(data4, bins=30, alpha=1, label="Dist 4")
plt.title("Histogramme comparatif")
plt.xlabel("Valeurs")
plt.ylabel("Fréquence")
plt.legend()
plt.show()
```

```python
# Bar chart 

import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D", "E"]
valeurs = [5, 7, 3, 8, 6]

plt.bar(categories, valeurs, color="blue", edgecolor="black")
plt.title("Diagramme en barres")
plt.xlabel("Catégories")
plt.ylabel("Valeurs")
plt.show()
```

```python
# Bar horizontale 

plt.barh(categories, valeurs, color="lightgreen", edgecolor="black")
plt.title("Diagramme en barres horizontales")
plt.xlabel("Valeurs")
plt.ylabel("Catégories")
plt.show()
```

```python
# Comparaisons de 2 séries 
import numpy as np

# Données
categories = ["A", "B", "C", "D"]
valeurs1 = [4, 7, 3, 5]
valeurs2 = [6, 2, 5, 8]

x = np.arange(len(categories))  # position des barres
largeur = 0.35  # largeur des barres

# Bar chart comparatif
plt.bar(x - largeur/2, valeurs1, largeur, label="Série 1", color="skyblue")
plt.bar(x + largeur/2, valeurs2, largeur, label="Série 2", color="orange")

plt.xticks(x, categories)
plt.title("Comparaison de deux séries")
plt.xlabel("Catégories")
plt.ylabel("Valeurs")
plt.legend()
plt.show()
```

```python
# Nuages de point de données 

import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = 2 * x + np.random.randn(50) * 0.1  # relation linéaire avec un peu de bruit

plt.scatter(x, y, marker="o", color="black")
plt.title("Nuage de points simple")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```

```python
# Nuage de points amélioré 

x = np.random.rand(100)
y = np.random.rand(100)
taille = np.random.rand(100) * 300   
couleur = np.random.rand(100)       

plt.scatter(x, y, s=taille,c=couleur, cmap="viridis", alpha=0.7, edgecolors="black")
plt.title("Nuage de points avec tailles et couleurs")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label="Valeur associée") 
plt.show()
```

```python
# Plusieurs groupe de nuage de point

# Groupe 1
x1 = np.random.normal(2, 0.5, 50)
y1 = np.random.normal(2, 0.5, 50)

# Groupe 2
x2 = np.random.normal(4, 0.5, 50)
y2 = np.random.normal(4, 0.5, 50)

plt.scatter(x1, y1, color="red", label="Groupe 1", alpha=0.7)
plt.scatter(x2, y2, color="blue", label="Groupe 2", alpha=0.7)
plt.title("Nuage de points par groupes")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

# Seaborn 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Jeu de données simple
np.random.seed(42)
df = pd.DataFrame({
    "x": np.arange(1, 21),
    "y": np.random.randint(10, 100, 20),
    "cat": np.random.choice(["A", "B", "C"], size=20),
    "valeurs": np.random.randn(20)
})

# STYLE GLOBAL du graphique 
sns.set_theme(style="whitegrid")
```

```python
# Exemples des différents style 
for style in ["darkgrid", "whitegrid", "dark", "white", "ticks"]:
    sns.set_theme(style=style)
    sns.histplot(np.random.randn(100), bins=20)
    plt.title(f"Style = {style}")
    plt.show()
```

```python
# Courbes 
sns.lineplot(x="x",y="y",data=df, marker="o", hue="cat")
plt.title("Courbes avec Seaborn (lineplot)")
plt.show()
```

```python
# Histogrammes 
sns.histplot(df["valeurs"], bins = 20, kde=True, color='grey', edgecolor='black')
plt.title("Histogramme avec Seaborn (histplot)")
plt.show()
```

```python
# Barre verticale 
sns.barplot(x="cat", y="y", data=df, estimator=np.mean, palette="viridis")
plt.title("Diagramme en barres verticales")
plt.show()
```
